# Visulization Recommender by Hariom Vyas

Upload data and it recommends you types of visulization and which columns to use for those visulization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Try 1

```python
def suggest_visualization(data):
    # Identify numerical and categorical columns
    numerical_cols = data.select_dtypes(include=['int', 'float']).columns
    categorical_cols = data.select_dtypes(include=['object']).columns

    # Analyze numerical columns
    numerical_graph = None
    numerical_cols_best = []
    numerical_cols_scores = {}

    for col in numerical_cols:
        # Perform scoring/ranking for numerical columns (you can customize this part based on your criteria)
        score = len(data[col].unique()) + data[col].count() / len(data[col].unique())
        numerical_cols_scores[col] = score

    if numerical_cols_scores:
        # Sort the numerical columns based on scores (higher score is better)
        numerical_cols_best = sorted(numerical_cols_scores, key=numerical_cols_scores.get, reverse=True)
        numerical_graph = 'bar'  # Assuming bar chart is a good default choice for numerical data

    # Analyze categorical columns
    categorical_graph = None
    categorical_cols_best = []
    categorical_cols_scores = {}

    for col in categorical_cols:
        # Perform scoring/ranking for categorical columns (you can customize this part based on your criteria)
        score = data[col].nunique() / data[col].count()
        categorical_cols_scores[col] = score

    if categorical_cols_scores:
        # Sort the categorical columns based on scores (higher score is better)
        categorical_cols_best = sorted(categorical_cols_scores, key=categorical_cols_scores.get, reverse=True)
        categorical_graph = 'bar'  # Assuming bar chart is a good default choice for categorical data

    # Return the suggested visualization types and best columns
    return {
        'numerical': {
            'graph': numerical_graph,
            'columns': numerical_cols_best
        },
        'categorical': {
            'graph': categorical_graph,
            'columns': categorical_cols_best
        }
    }
```

```python
# Example usage
data = pd.read_csv('countries.csv')  # Replace 'data.csv' with your data source

suggested_visualizations = suggest_visualization(data)

numerical_graph = suggested_visualizations['numerical']['graph']
numerical_cols_best = suggested_visualizations['numerical']['columns']

categorical_graph = suggested_visualizations['categorical']['graph']
categorical_cols_best = suggested_visualizations['categorical']['columns']

print('Numerical data:')
print('Suggested graph:', numerical_graph)
print('Best columns:', numerical_cols_best)

print('\nCategorical data:')
print('Suggested graph:', categorical_graph)
print('Best columns:', categorical_cols_best)
```

```python
def suggest_visualization(data):
    # Identify numerical, categorical, and date columns
    numerical_cols = data.select_dtypes(include=['int', 'float']).columns
    categorical_cols = data.select_dtypes(include=['object']).columns
    date_cols = data.select_dtypes(include=['datetime']).columns

    # Analyze numerical columns
    numerical_graph = None
    numerical_cols_best = []
    numerical_cols_scores = {}

    for col in numerical_cols:
        # Perform scoring/ranking for numerical columns (you can customize this part based on your criteria)
        score = len(data[col].unique()) + data[col].count() / len(data[col].unique())
        numerical_cols_scores[col] = score

    if numerical_cols_scores:
        # Sort the numerical columns based on scores (higher score is better)
        numerical_cols_best = sorted(numerical_cols_scores, key=numerical_cols_scores.get, reverse=True)
        numerical_graph = 'bar'  # Assuming bar chart is a good default choice for numerical data

    # Analyze categorical columns
    categorical_graph = None
    categorical_cols_best = []
    categorical_cols_scores = {}

    for col in categorical_cols:
        # Perform scoring/ranking for categorical columns (you can customize this part based on your criteria)
        score = data[col].nunique() / data[col].count()
        categorical_cols_scores[col] = score

    if categorical_cols_scores:
        # Sort the categorical columns based on scores (higher score is better)
        categorical_cols_best = sorted(categorical_cols_scores, key=categorical_cols_scores.get, reverse=True)
        categorical_graph = 'bar'  # Assuming bar chart is a good default choice for categorical data

    # Analyze date columns
    date_graph = None
    date_cols_best = []
    date_cols_scores = {}

    for col in date_cols:
        # Perform scoring/ranking for date columns (you can customize this part based on your criteria)
        score = data[col].nunique() / data[col].count()
        date_cols_scores[col] = score

    if date_cols_scores:
        # Sort the date columns based on scores (higher score is better)
        date_cols_best = sorted(date_cols_scores, key=date_cols_scores.get, reverse=True)
        date_graph = 'line'  # Assuming line chart is a good default choice for date data

    # Return the suggested visualization types and best columns
    return {
        'numerical': {
            'graph': numerical_graph,
            'columns': numerical_cols_best
        },
        'categorical': {
            'graph': categorical_graph,
            'columns': categorical_cols_best
        },
        'date': {
            'graph': date_graph,
            'columns': date_cols_best
        }
    }
```

```python
def create_visualizations(data, suggested_visualizations):
    for col in suggested_visualizations['numerical']['columns']:
        if suggested_visualizations['numerical']['graph'] == 'bar':
            data[col].value_counts().plot(kind='bar', figsize=(10, 6))
            plt.title(f'Bar chart for {col}')
            plt.xlabel(col)
            plt.ylabel('Count')
            plt.show()

    for col in suggested_visualizations['categorical']['columns']:
        if suggested_visualizations['categorical']['graph'] == 'bar':
            data[col].value_counts().plot(kind='bar', figsize=(10, 6))
            plt.title(f'Bar chart for {col}')
            plt.xlabel(col)
            plt.ylabel('Count')
            plt.show()

    for col in suggested_visualizations['date']['columns']:
        if suggested_visualizations['date']['graph'] == 'line':
            data[col].value_counts().sort_index().plot(kind='line', figsize=(10, 6))
            plt.title(f'Line chart for {col}')
            plt.xlabel(col)
            plt.ylabel('Count')
            plt.show()
```

```python
# Example usage
data = pd.read_csv('countries.csv')  # Replace 'data.csv' with your data source

suggested_visualizations = suggest_visualization(data)
create_visualizations(data, suggested_visualizations)
```

## Try 4 (With Seaborn)

```python
def suggest_visualizations(data):
    suggested_visualizations = {}

    # Identify numerical columns
    numerical_cols = data.select_dtypes(include=['int', 'float']).columns
    suggested_visualizations['numerical'] = {
        'graphs': ['histplot', 'boxplot', 'violinplot'],
        'columns': numerical_cols.tolist()
    }

    # Identify categorical columns
    categorical_cols = data.select_dtypes(include=['object']).columns
    suggested_visualizations['categorical'] = {
        'graphs': ['countplot', 'barplot'],
        'columns': categorical_cols.tolist()
    }

    # Identify date columns
    date_cols = data.select_dtypes(include=['datetime']).columns
    suggested_visualizations['date'] = {
        'graphs': ['lineplot', 'scatterplot'],
        'columns': date_cols.tolist()
    }

    return suggested_visualizations
```

```python
def create_visualizations(data, suggested_visualizations):
    for dtype, visualization in suggested_visualizations.items():
        for col in visualization['columns']:
            for graph_type in visualization['graphs']:
                plt.figure(figsize=(10, 6))
                
                if graph_type == 'histplot':
                    sns.histplot(data=data, x=col)
                    plt.title(f'Histogram for {col}')
                    plt.xlabel(col)
                    plt.ylabel('Count')
                
                elif graph_type == 'boxplot':
                    sns.boxplot(data=data, x=col)
                    plt.title(f'Box Plot for {col}')
                    plt.xlabel(col)
                    plt.ylabel('Value')
                
                elif graph_type == 'violinplot':
                    sns.violinplot(data=data, x=col)
                    plt.title(f'Violin Plot for {col}')
                    plt.xlabel(col)
                    plt.ylabel('Value')
                
                elif graph_type == 'countplot':
                    sns.countplot(data=data, x=col)
                    plt.title(f'Count Plot for {col}')
                    plt.xlabel(col)
                    plt.ylabel('Count')
                
                elif graph_type == 'barplot':
                    sns.barplot(data=data, x=col, y=data[col].value_counts())
                    plt.title(f'Bar Plot for {col}')
                    plt.xlabel(col)
                    plt.ylabel('Count')
                
                elif graph_type == 'lineplot':
                    sns.lineplot(data=data, x=col, y=data[col].value_counts())
                    plt.title(f'Line Plot for {col}')
                    plt.xlabel(col)
                    plt.ylabel('Count')
                
                elif graph_type == 'scatterplot':
                    sns.scatterplot(data=data, x=col, y=data[col].value_counts())
                    plt.title(f'Scatter Plot for {col}')
                    plt.xlabel(col)
                    plt.ylabel('Count')

                plt.show()
```

```python
# Example usage
data = pd.read_csv('countries.csv')  # Replace 'data.csv' with your data source

suggested_visualizations = suggest_visualizations(data)
create_visualizations(data, suggested_visualizations)
```
